```python
import numpy as np
import pandas as pd
import networkx as nx
from networkx.drawing.nx_agraph import graphviz_layout
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
G = nx.Graph()
G.add_node("Party Leader", type='user')
for i in range(4):
    G.add_node("Tweet {}".format(i+1),type='tweet')
    G.add_edge("Party Leader", "Tweet {}".format(i+1))
G.add_node("General User",type='retweet')
G.add_edge("General User","Tweet 1")
G.add_edge("General User","Tweet 3")
```

```python
colors = []
for node in G.nodes():
    attributes = G.nodes[node]
    color = 'red'
    if attributes['type'] == 'retweet':
        colors.append('#79BFD3')
    elif attributes['type'] == 'tweet':
        colors.append("#006816")
    elif attributes['type'] == 'user':
        colors.append('red') 
print(colors)
```

```python
plt.figure(figsize=(10,10))
plt.title("Example Engagement Graph",fontdict={'fontsize':20})
pos = graphviz_layout(G, prog="dot")
nx.draw(G, pos,
                node_color=colors,
                with_labels=True,
                font_size=16,
                node_size=200,
                width=2,
                arrows=False
                )
plt.tight_layout()
plt.autoscale()
plt.savefig("../visualizations/ex_engagement_graph3.png",dpi=300, bbox_inches='tight')
plt.show()
```
